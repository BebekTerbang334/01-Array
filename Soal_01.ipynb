Sistem sederhana pengelolaan nilai mahasiswa menggunakan array.

1. Input 10 nilai

```python
nilai=[]
for i in range(10):
    angka = int(input("Masukkan Nilai Mahasiswa : "))
    nilai.append(angka)

print("Nilai Mahasiswa : ", nilai)

```

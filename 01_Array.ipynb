Mendefinisikan list kosong untuk menyimpan `nilai` Mahasiswa



```python
nilai=[]
```

Melakukan `import matplotlib.pyplot as plt` untuk menunjukkan grafik

```python
import matplotlib.pyplot as plt
```

Menggunakan perulangan `for` untuk meminta input nilai Mahasiswa dari user sebanyak 10 kali dan menampilkannya menggunakan `print("Nilai Mahasiswa : ", nilai)`

```python
for i in range(10):
    angka = int(input("Masukkan Nilai Mahasiswa : "))
    nilai.append(angka)

print("Nilai Mahasiswa : ", nilai)
```

Menampilkan nilai tertinggi dan terendah mahasiswa menggunakan `max(nilai)` dan `min(nilai)`

```python
print("Nilai Tertinggi : ", max(nilai))
print("Nilai Terendah : ", min(nilai))
```

Menampilkan nilai rata-rata mahasiswa menggunakan `sum(nilai)/len(nilai)`

`sum(nilai)` untuk menjumlahkan nilai-nilai yang ada dalam array

`len(nilai)` untuk menampilkan total nilai dalam array

```python
print("Nilai Rata-rata : ", sum(nilai)/len(nilai))
```

Menggunakan `for` untuk perulangan pada setiap elemen dalam array/list `nilai`

Menggunakan `if` digunakan untuk mengecek kondisi, jika kondisi >= 60 maka `lulus += 1`

Menghitung `tidak_lulus` dengan mengurangi total nilai ( `len(nilai)`) dengan total `lulus`

Lalu menampilkan jumlah mahasiswa yang lulus dan tidak lulus menggunakan `print`

```python
lulus = 0
for n in nilai:
    if n >= 60:
        lulus += 1
tidak_lulus = len(nilai) - lulus

print("Jumlah Mahasiswa yang lulus : ", lulus)
print("Jumlah Mahasiswa yang tidak lulus : ", tidak_lulus)
```

Menampilkan grafik untuk menggambarkan nilai tertinggi dan terendah dan grafik untuk menggambarkan data kelulusan Mahasiswa

```python
plt.bar(["Tertinggi", "Terendah"], [max(nilai), min(nilai)])
plt.title("Grafik Nilai Tertinggi dan Terendah")
plt.ylabel("Nilai")
plt.ylim([0, 100])
plt.show()

plt.bar(["Lulus", "Tidak Lulus"], [lulus, tidak_lulus])
plt.title("Grafik Kelulusan Mahasiswa")
plt.ylabel("Jumlah Mahasiswa")
plt.ylim([0, 10])
plt.show()
```
